```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
url = 'https://www.abet.org/accreditation/accreditation-criteria/criteria-for-accrediting-engineering-programs-2025-2026/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
#print(soup)
```

```python
download_link = soup.find('a', string=lambda text: text and 'Download the Criteria' in text)

if download_link and download_link.get('href'):
    pdf_url = urljoin(url, download_link['href'])
    print(f"PDF URL: {pdf_url}")
    
    # Download the PDF
    response = requests.get(pdf_url, allow_redirects=True)
    
    with open('ABET_2025-2026_EAC_Criteria.pdf', 'wb') as f:
        f.write(response.content)
    
    print("PDF downloaded successfully!")
else:
    print("Download link not found")
```

```python
courses = soup.find_all('li', class_='block-accordion-item')
for item in courses:
    # Main title
    main_title = item.find("h4", class_="accordion-header").get_text(strip=True)
    if(main_title == 'Systems and Similarly Named Engineering Programs'):
        print(item)

```

```python
CSE_data = [title.text.strip() for title in item]
CSE_data = [text for text in CSE_data if text]
print(CSE_data)
```

```python
CSE_wtxt = open("CSE_datafile.txt", "w")

for i in CSE_data:
    CSE_wtxt.write( i + "\n")

CSE_wtxt.close()
```

```python
CSE_rtxt = open("CSE_datafile.txt", "r+")

nummy = 0
for i in CSE_rtxt:
    nummy += 1
    print(str(nummy) + ". " + i.strip())
```

```python
# General Criteria for Baccalaureate Level Programs
start = soup.find('h2', id='1')
print(start)
if start:
    content = []
    sibling = start.find_next_sibling()

    while sibling and not (sibling.name == "h2" and sibling.get("id") == "2"):
        content.append(str(sibling))
        sibling = sibling.find_next_sibling()

    print(BeautifulSoup("".join(content), "html.parser").prettify())
```
