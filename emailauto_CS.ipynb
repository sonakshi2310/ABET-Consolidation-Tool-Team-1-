```python
#Import
import os
import smtplib #for email sending
from email.message import EmailMessage
import difflib
from dotenv import load_dotenv
load_dotenv()

#SendGrid import (demo purpose)
%pip install sendgrid
from sendgrid import SendGridAPIClient
from sendgrid.helpers.mail import Mail
```

```python
#Configurations

OLD_FILE = "oldCS_datafile.txt"
NEW_FILE = "CS_datafile.txt"
# if using Google SMTP
#SMTP_SERVER = "smtp.gmail.com"
#SMTP_PORT = 587
#EMAIL_PASSWORD = os.getenv("EMAIL_PASSWORD")  # Make sure to set this environment variable
```

```python
#check if files exist
if not os.path.exists(OLD_FILE):
    print(f"{OLD_FILE} does not exist. Creating a new one.")
    exit(1)

if not os.path.exists(NEW_FILE):
    print(f"{NEW_FILE} does not exist. Creating a new one.")
    exit(1)

# Can use this when sure both files are UTF-8 encoded
with open(OLD_FILE, 'r') as old_file, open(NEW_FILE, 'r') as new_file:
    old_data = old_file.readlines()
    new_data = new_file.readlines()


# Read files with error handling for encoding issues in case needed
# try:
#     with open(OLD_FILE, 'r', encoding='utf-8') as old_file:
#         old_data = old_file.readlines()
# except UnicodeDecodeError:
#     # Try different encodings if UTF-8 fails
#     try:
#         with open(OLD_FILE, 'r', encoding='latin-1') as old_file:
#             old_data = old_file.readlines()
#     except Exception as e:
#         print(f"Error reading {OLD_FILE}: {e}")
#         exit(1)

# try:
#     with open(NEW_FILE, 'r', encoding='utf-8') as new_file:
#         new_data = new_file.readlines()
# except UnicodeDecodeError:
#     # Try different encodings if UTF-8 fails
#     try:
#         with open(NEW_FILE, 'r', encoding='latin-1') as new_file:
#             new_data = new_file.readlines()
#     except Exception as e:
#         print(f"Error reading {NEW_FILE}: {e}")
#         exit(1)

# Can add lineterm parameter if needed
diff = list(difflib.unified_diff(old_data, new_data, fromfile='oldCS_datafile.txt', tofile='CS_datafile.txt'))


if diff:
    #demo print diff
    diff_text = ''.join(diff)
    print("Differences found:")
    print(diff_text)
    

    #Prepare email demonstration 
    message = Mail(
        from_email=os.getenv("EMAIL_SENDER"),
        to_emails=os.getenv("EMAIL_RECEIVER"),
        subject="CS Datafile Changes Detected",
        plain_text_content=f"The following changes were detected between the old and new CS data files:\n\n{diff_text}"
    )

    sg = SendGridAPIClient(os.getenv("SENDGRID_API_KEY"))
    try:
        response = sg.send(message)
        print(f"Email sent with status code {response.status_code}")
    except Exception as e:
        print(f"Error sending email: {e}")

#How to use:
#Set up .env file with necessary info (see .env file for reference)
#Create SendGrid account and get API key

#What to do:
#Change SendGrid account to using ASU email address
#Get API key and set up .env file with necessary info
#Update the contect of the old_datafile.txt with the content of current CS_datafile.txt after verifying the changes
#Automate this script to run at desired intervals
```
